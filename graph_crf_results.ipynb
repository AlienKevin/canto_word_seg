```python
import matplotlib.pyplot as plt

# Increase font size
plt.rcParams.update({'font.size': 20})

# List of model names
models = ['10%', '20%', '30%', '40%', '50%', '60%', '70%', '80%', '90%']

# Corresponding precision scores
precision = [89.91, 91.28, 91.75, 92.19, 92.26, 92.45, 92.61, 92.86, 93.03]

# Corresponding recall scores
recall = [90.17, 91.48, 91.85, 92.37, 92.48, 92.61, 92.60, 92.81, 93.03]

# Corresponding F1 scores
f1 = [90.04, 91.38, 91.80, 92.28, 92.37, 92.53, 92.61, 92.83, 93.03]

# Corresponding accuracy scores
accuracy = [94.11, 94.97, 95.28, 95.55, 95.57, 95.64, 95.67, 95.87, 95.95]

plt.figure(figsize=(12, 8))

# Plotting all metrics on the same graph
plt.plot(models, precision, marker='o', label='Precision')
plt.plot(models, recall, marker='o', label='Recall')
plt.plot(models, f1, marker='o', label='F1')
plt.plot(models, accuracy, marker='o', label='Accuracy')

plt.title('Metrics of CRF Models')
plt.xlabel('Models')
plt.ylabel('Score (%)')
plt.legend()
plt.grid(True)
plt.show()
```
